### Data Cleaning

```python
# Checking the data type of the dataframe.
df.info()
```

```python
# Changing the datatype of Date and Updated_On from object to DateTime
df['Date'] = pd.to_datetime(df['Date'])
print(f"Date Datatype: {df['Date'].dtype}")
df["Updated On"] = pd.to_datetime(df['Updated On'])
print(f"Updated On Datatype: {df['Updated On'].dtype}")
```

```python
# Checking to see if the change has been effected.
df.info()
```

```python
# Checking for null values
df.isna().sum()
```

```python
# Dropping the columns with null values in the dataframe
df.dropna(axis=0, inplace=True)
```

```python
#Checking if the columns with null value has been dropped
df.isna().sum()
```

```python
# Checking for the number of rows and columns contained in the data
df.shape
```

```python
# Renaming of columns from space separated to underscore
old_column = [x for x in df.columns if ' ' in x]
df = df.rename(columns= {x: x.replace(' ', '_') for x in old_column})
df
```

```python
df.columns
```

```python
# Checking to see if there are duplicate entries in the dataframe
df.duplicated().sum()
```

```python
# Checking the most important column for duplicate entries
df['Case_Number'].duplicated().sum()
```

```python
# Here, i dropped the ID column to see if the number of duplicate entries will reduce
df.drop('ID', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Checking again for duplicate entries
df.duplicated().sum()
```

```python
# Finally dropping the all the duplicate entries
df = df.drop_duplicates()
```

```python
# Checking to see if the duplicaated dropped has been effected
df.duplicated().sum()
```

### Descriptive Analysis (Numerical Variables)

```python
df.describe()
```

```python
# Sorting the primary type of offence 
df["Primary_Type"].value_counts()
```

```python
# Checking to know which of these crimes were arrest made
df['Arrest'].value_counts()
```

From the output above, we can see that not all the cases entered were solved. The True output shows that the cases were solved abd the False output shows that the cases were not solved.

```python
df_arrest = df["Arrest"].value_counts()
plt.title('Solved Cases vs Unsolved Cases', size = 10, loc ='center', fontweight = 'bold')
plt.ylabel(' ')
sns.barplot(df_arrest)
```

##### From the graph above, we can see that that the False arrest is more than the True arrest.

```python
# Checking to see the district where crime is more prevalent
df['District'].value_counts()
```

```python
# Plotting the graph of the unsolved and solved cases of the District
df_district = df["District"].value_counts()
plt.title('Solved Cases vs Unsolved Cases', size = 10, loc ='center', fontweight = 'bold')
plt.ylabel(' ')
plt.bar(x = df_district.index, height = df_district.values)
```

The observation from the bar chart above shows the district in which crimes are prevalent

```python
# Grouping the data by year 
df_2006 = df[(df['Year'] == 2001) | (df['Year'] == 2002) | (df['Year'] == 2003) | (df['Year'] == 2004) | (df['Year'] == 2005) | (df['Year'] == 2006)]['Year'].value_counts()
df_2012 = df[(df['Year'] == 2007) | (df['Year'] == 2008) | (df['Year'] == 2009) | (df['Year'] == 2010) | (df['Year'] == 2011) | (df['Year'] == 2012)]['Year'].value_counts()
df_2018 = df[(df['Year'] == 2013) | (df['Year'] == 2014) | (df['Year'] == 2015) | (df['Year'] == 2016) | (df['Year'] == 2017) | (df['Year'] == 2018)]['Year'].value_counts()
df_2023 = df[(df['Year'] == 2019) | (df['Year'] == 2020) | (df['Year'] == 2021) | (df['Year'] == 2022) | (df['Year'] == 2023)]['Year'].value_counts()
```

```python
df_2006
```

```python
# Bar plot for the grouped years
fig, axes = plt.subplots(2, 2, figsize = (16, 10))
plt.suptitle('2001 - 2023 District Unsolved Crime Distribution', fontweight = 'bold', size = 14)
sns.barplot(y = df_2006.values, x = df_2006.index, ax=axes[0,0])
sns.set_palette('viridis')
axes[0,0].set_title('2001 - 2006 Unsolved Crime Distribution')

sns.barplot(y = df_2012.values, x = df_2012.index, ax=axes[0,1])
axes[0,1].set_title('2007 - 2012 Unsolved Crime Distribution')

sns.barplot(y = df_2018.values, x = df_2018.index, ax=axes[1,0])
axes[1,0].set_title('2013 - 2018 Unsolved Crime Distribution')

sns.barplot(y = df_2023.values, x = df_2023.index, ax=axes[1,1])
axes[1,1].set_title('2019 - 2023 Unsolved Crime Distribution')
```
